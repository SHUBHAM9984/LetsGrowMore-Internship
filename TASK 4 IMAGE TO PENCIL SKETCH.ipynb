```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread(r"/content/drive/MyDrive/RAM JI.jpeg")
cv2_imshow(image)
```

```python
#Converting the original image to greyscale
```

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_image)
```

```python
#Inverting the greyscale image
```

```python
inverted_image = 255 - grey_image
cv2_imshow(inverted_image)
```

```python
#Blurring the image by using the Gaussian Function in OpenCV
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(blurred)
```

```python
#PENCIL SKETCH
```

```python
def pencil_sketch(a, b):
  return cv2.divide(a, 255 - b, scale=256)
pencil_sketch=pencil_sketch(grey_image,blurred)
cv2_imshow(pencil_sketch)
```
