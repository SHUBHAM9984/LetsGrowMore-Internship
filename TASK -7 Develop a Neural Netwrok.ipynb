# IMPORTING REQUIRED LIBARY 

```python
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential 
from keras.layers.core import Dense, Activation
from keras.utils import np_utils
```

# Loading the MNIST handwritten digit data

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
n = 40  
plt.figure(figsize=(40, 4))
for i in range(n):
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
plt.close()
```

```python
print("Previous x_train shape: {} \nPrevious y_train shape:{}".format(x_train.shape, y_train.shape))
x_train = x_train.reshape(60000, 784)     
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')     
x_test = x_test.astype('float32')     
x_train /= 255    
x_test /= 255
classes = 10
y_train = np_utils.to_categorical(y_train, classes)     
y_test = np_utils.to_categorical(y_test, classes)
print("New x_train shape: {} \nNew y_train shape:{}".format(x_train.shape, y_train.shape))
```

```python
#Setting parameters
input_size = 784
batch_size = 200   
hidden1 = 400
hidden2 = 20
epochs = 2
```

```python
#Building Model
model = Sequential()     
model.add(Dense(hidden1, input_dim=input_size, activation='relu'))

model.add(Dense(hidden2, activation='relu'))
model.add(Dense(classes, activation='softmax')) 

model.compile(loss='categorical_crossentropy', 
metrics=['accuracy'], optimizer='sgd')
model.summary()
```

```python
model.fit(x_train, y_train, batch_size=batch_size, epochs=100, verbose=2)
```

```python
#TESTING THE MODEL
```

```python
score = model.evaluate(x_test, y_test, verbose=1)
print('\n''Test accuracy:', score[1])
mask = range(0,40)
x_valid = x_test[mask]
predict_y=model.predict(x_valid) 
classes_y=np.argmax(predict_y,axis=1)
print(classes_y)

# display original
plt.figure(figsize=(40, 4))
for i in range(n):
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_valid[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
plt.close()
```
